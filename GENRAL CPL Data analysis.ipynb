```python
data["Toss"].value_counts()
```

```python
field = data["Toss"].value_counts().to_frame()
field
```

```python
# CHecking Sum of tosses For top 4 teams
jamica   = field.loc[field.index.str.contains('Jamaica'), "Toss"].sum()
guyana   = field.loc[field.index.str.contains('Guyana'), "Toss"].sum()
st_Kitts = field.loc[field.index.str.contains('St Kitts'), "Toss"].sum()
trinbago = field.loc[field.index.str.contains('Trinbago'), "Toss"].sum()
print(f"JAMICA : {jamica} \nGUYANA : {guyana} \nST_KITTS : {st_Kitts} \nTRINBAGO:{trinbago}")
```

```python
toss_winner = pd.DataFrame({
    "jamica"   : field.loc[field.index.str.contains('Jamaica'), "Toss"],
    "guyana"   : field.loc[field.index.str.contains('Guyana'), "Toss"],
    "st_Kitts" : field.loc[field.index.str.contains('St Kitts'), "Toss"],
    "trinbago" : field.loc[field.index.str.contains('Trinbago'), "Toss"]
}
)
```

```python
toss_winner
```

```python
# Shifting the values By rows
toss_winner["jamica"] = toss_winner["jamica"].shift(-2) # Two rows
toss_winner["st_Kitts"] = toss_winner["st_Kitts"].shift(-4) # Fours rows
toss_winner["trinbago"] = toss_winner["trinbago"].shift(-6) # Six rows
```

```python
toss_winner = toss_winner.rename(index = {
    "Guyana Amazon Warriors opted for bat" : "BAT",
    "Guyana Amazon Warriors opted for field" : "BOWL"
})
```

```python
# After renaming the first two index and dropping irrelevant rows
toss_winner = toss_winner.drop(index=toss_winner.index[-6:])
```

```python
toss_winner
```

```python
# We use melt function to plot this graph where hue is set as index , columns = COuntry and values are set as Count
melted_data = pd.melt(toss_winner.reset_index(), id_vars='index' , var_name='City', value_name='Count')

# toss_winner.reset_index() create new column index with values BAT , BOWL
# id_vars='index' tells index columns is treated as identifier

    
ax = sns.catplot(x='City', y='Count', hue='index', data=melted_data, kind='bar', palette='Set1')

# Add data labels to the bars
for p in ax.ax.patches:
    ax.ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
ax.set(xlabel='City', ylabel='Count')
plt.style.use('cyberpunk')
plt.show()
```

```python
# Winning Counts and toss sum of top 7 most winning Teams 
relation = {"winner": [49,41,39,30,27,22,14,22],
            "toss":   [39,39,36,23,37,26,17,12]
           }
relation = pd.DataFrame(relation  , index = [1,2,3,4,5,6,7,8])
```

```python
mpl_style(dark=True)
sns.lmplot(x ='winner', y ='toss', data = relation ,fit_reg=True)
plt.grid(linewidth=0.175)
plt.show()
```
